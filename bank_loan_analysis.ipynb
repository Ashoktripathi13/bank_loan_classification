#  handling nan values

```python
df.isna().sum()
```

```python
online_median=df['Online'].median()
income_median=df['Income'].median()
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Home Ownership'] = df['Home Ownership'].astype('category')
```

```python
gender_mode=df['Gender'].mode()[0]
home_mode=df['Home Ownership'].mode()[0]
```

```python
df['Income'].fillna(value=income_median,inplace=True)
df['Online'].fillna(value=online_median,inplace=True)
```

```python
df['Gender'].fillna(value=gender_mode,inplace=True)
df['Home Ownership'].fillna(value=home_mode,inplace=True)
```
